```python
import pandas as pd
import os
data_1 =[]

for filename in os.listdir('C:/Users/i503649/Desktop/266/project/sentiments/sen'):
#     print(filename)
    with open(os.path.join('C:/Users/i503649/Desktop/266/project/sentiments/sen', filename), 'r',encoding="utf8") as f:
        data = f.readlines()
#          print(data)
    data_1.append(data)

for i in range(0,len(data_1)):
# remove the trailing "\n" from each line
    if i == 0:
        data_1[i] = map(lambda x: x.rstrip(), data_1[i])
    #     print(data_1[i])
        data_json_str = '[' + ','.join(data_1[i]) + ']'
        data_df = pd.read_json(data_json_str)


#         print(data_df_1)
    else:
        data_1[i] = map(lambda x: x.rstrip(), data_1[i])
    #     print(data_1[i])
        data_json_str = '[' + ','.join(data_1[i]) + ']'
        data_df_2 = pd.read_json(data_json_str)
        data_df = pd.concat([data_df, data_df_2])
```

```python
df = data_df_1.rename(index=str, columns={"sentiment_score": "Polarity", "text": "Tweet", "sentiment": "Sentiment"})
```

```python
df["tweet_lower"] = df['Tweet'].str.lower()
```

```python
df['tweet_lower']
```

```python
PlayerSentiment = pd.DataFrame(columns= ["Player", "Tweet", "Polarity", "Sentiment"])
```

```python
def PlayerSentimentDef(PlayerName, SearchString):
    global PlayerSentiment
    print(PlayerName)
    print(SearchString)
    playerDF = df.loc[df.tweet_lower.str.contains(SearchString)]
    playerDF.head()
    playerDF["Player"] = PlayerName
    playerDF = playerDF[ ["Player", "Tweet", "Polarity", "Sentiment"]]
    PlayerSentiment = PlayerSentiment.append(playerDF)
```

```python
PlayerSentimentDef("MS Dhoni", "dhoni")
PlayerSentimentDef("SK Raina", "raina")
PlayerSentimentDef("R Jadeja", "jadeja")
PlayerSentimentDef("S Dhawan", "dhawan")
PlayerSentimentDef("V Kohli", "kohli")
PlayerSentimentDef("Ch Gayle", "gayle")
PlayerSentimentDef("David Warner", "warner")
PlayerSentimentDef("Rohit Sharma", "rohit")
PlayerSentimentDef("ABD", "abd")
PlayerSentimentDef("Yusuf Pathan", "yusuf")
```

```python
PlayerSentiment = PlayerSentiment.drop_duplicates()
PlayerSentiment = PlayerSentiment.loc[PlayerSentiment.Polarity != 0]
```
